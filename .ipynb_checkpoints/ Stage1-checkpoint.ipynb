## filling of nan value in column

```python
data['C1) Funding FY2008'].fillna(0,inplace=True)
```

```python
data['C1) Funding FY2008'].isnull().sum()
```

```python
data['C2) Funding FY2009'].fillna(0,inplace=True)
```

```python
data['C1) Funding FY2008'].isnull().sum()
```

### calculation

```python
data['Growth']=""
```

```python
for i in range(len(data)):
    # dealing with denominator 0
    if data['C1) Funding FY2008'][i]==0.0: 
        if  data['C2) Funding FY2009'][i]==0.0:
            data['Growth'][i]=0
        elif data['C2) Funding FY2009'][i] <0.0:
             data['Growth'][i] = -100.0
        else:
             data['Growth'][i] = 100.0
    else:
        data['Growth'][i] = ((data['C2) Funding FY2009'][i] - data['C1) Funding FY2008'][i]) / data['C1) Funding FY2008'][i]) * 100
     
```

```python
data.head(1)
```

### assigning target 

```python
data['target'] = (data['Growth'] >= 0).astype(int)
```

```python
data.to_csv("data/ml_stem_analysis.csv",index=False)
```
